
For the analysis, two actions were requested:

The sum of all "Ratatouille" sales during the period;
The frequency of payment methods for "Ratatouille".

```python
sum_ratatoullie = dados.query("produto == 'Ratatouille'")['valor'].sum()
sum_ratatoullie
```

```python
freq_pay_mtd_ratatoullie = dados.query("produto == 'Ratatouille'")['metodo_pagamento'].value_counts()
freq_pay_mtd_ratatoullie
```

```python
print(f"As vendas de Ratatouille foram de R${sum_ratatoullie}")
```

```python
freq_pay_mtd_ratatoullie = freq_pay_mtd_ratatoullie.reset_index()
freq_pay_mtd_ratatoullie.columns = ['metodo_pagamento', 'frequencia']
freq_pay_mtd_ratatoullie
```

```python
freq_pay_mtd_ratatoullie.set_index('metodo_pagamento', inplace = True)
```

```python
freq_pay_mtd_ratatoullie['frequencia'].plot(kind='bar', figsize=(8, 5))
```
