```python
import pandas as pd
```

```python
vendas = pd.read_csv("/content/drive/MyDrive/Alura/Analise-de-notas-filmes/Homeworks/serenatto_2sem_2023.csv")
```

```python
# filter data in a column
## Return the only lines that the word was tiramisu
vendas.query('produto == "Tiramisù"').head()
```

```python
# filter two or more
## we need to match a logic symbols and operators for this
### in this case we filtered only products that costs >10 and selected all payments metods without a "dinheiro"
vendas.query('valor > 10 and metodo_pagamento != "Dinheiro"')
```

```python
# Query as a operator @
## used top refers an extern var

produtos = ['Café au lait', 'Espresso', 'Cappuccino']
vendas.query('produto in @produtos and metodo_pagamento == "PIX"')
```
