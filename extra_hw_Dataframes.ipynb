#Data frames and Series
##  goal of this homework is a exercise focusing on two concepts: DataFrame and Series. The task is to create a DataFrame and Series, and perform an exploratory analysis

```python
import pandas as pd
```

```python
# make a Data frame
## a small dict

dados = {'Nome':['Thiago', 'Matheus', 'João'],
         'Idade':[20, 25, 50],
         'Salario':[700, 3000, 20000]}
```

```python
df = pd.DataFrame(dados)
df
```

```python
df.info()  # Explore: timestamps, var type, null data and memory
```

```python
df['Salario'].mean()
```

```python
df['Salario'].median()
```

```python
# make a Series
```

```python
idades = pd.Series([20, 25, 50])
idades
```

```python
# a small difference between DataFrame and Series
# although a DataFrame is a collection of Series, when we explore a DataFrame, there is an additional line that contains the name of each Series
```

```python
idades.index
```

```python
idades.sort_values() # sort in ascending order
```

```python
idades.sort_values(ascending = False) # sort in descending order
```

```python
# about df is a set of series
nomes = pd.Series(["Ana", "Bruno", "Carla", "Diego", "Elisa"], name="nome")
idades = pd.Series([25, 34, 18, 40, 29], name="idade")
cidades = pd.Series(["São Paulo", "Rio de Janeiro", "Belo Horizonte", "Curitiba", "Porto Alegre"], name="cidade")
```

```python
df = pd.DataFrame({"Nomes":nomes,
                   "Idades":idades,
                   "Cidades":cidades})
df
```
